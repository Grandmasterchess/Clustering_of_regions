Создадим функцию для изменения наименования в строке и преобразование в список

```python
def replace_Tur(Characters):
    Characters = Characters.replace(",обл."," обл.")
    Characters = Characters.replace(",АО"," АО")
    Characters = Characters.replace(",край"," край")
    Characters = Characters.replace(",Эл"," Эл")
    Characters = Characters.replace(",Осетия"," Осетия")
    Characters = Characters.split(',')
    Characters = list(filter(None, Characters))
    return Characters
```

```python
Spisok = replace_Tur(Stroka)
my_list = Spisok[:-45]
print(len(my_list))
```

Сделаем из списка датафрейм

```python
df = pd.DataFrame(np.array(my_list).reshape(85, 23)).add_prefix("col")
# заменим '-' на Nan
mask = df.applymap(lambda x: x == '-')
df = df.replace('-', np.nan).fillna(method='ffill')
# переведём в тип float
df1 = df['col0']
df.drop(['col0'], axis=1, inplace = True)
for col in df.columns:
    df[col] = df[col].astype(float)
# умножим на 1000, так как в таблице приведены данные в тыс. чел
df = df *1000
df = pd.concat([df1, df], axis=1)
df.head(3)
```

Укажим наименования столбцов и отсортируем года в обратном порядке

```python
df.rename(columns={df.columns[i]: f'year{i-2022}' for i in range(len(df.columns))}, inplace=True)
df.rename(columns=lambda x: x.replace('year-', ''), inplace=True)
# отсортируем столбцы
df = df.reindex(sorted(df.columns), axis=1)
# переставим столбец регион в начало
df['2000'] = df['2022']
df.rename(columns={'2000': 'регион'}, inplace=True)
df.drop(['2022'], axis=1, inplace = True)
df.head(3)
```
